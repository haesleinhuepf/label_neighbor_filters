```python
import numpy as np
import pyclesperanto_prototype as cle
from skimage.io import imsave, imread
import matplotlib.pyplot as plt

cle.select_device("RTX")
```

```python
labels = cle.push(imread("lund_labels.tif"))

print(labels.shape)

cle.imshow(labels, labels=True)
```

```python
def show_with_colorbar(image):
    projection = cle.maximum_z_projection(image)
    
    fig, (ax1) = plt.subplots(figsize=(13, 5), ncols=1)

    pos = ax1.imshow(np.asarray(projection), cmap='hot', interpolation='none', vmin=0, vmax=2.9)
    fig.colorbar(pos, ax=ax1)
    

extension_ratio_map = cle.extension_ratio_map(labels)

show_with_colorbar(extension_ratio_map)
```
